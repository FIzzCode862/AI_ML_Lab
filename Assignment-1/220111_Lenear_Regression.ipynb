```python
X = df.drop(columns=["price"], axis=1)
y = df["price"]
```

```python
X
```

```python
y
```

```python
X_one_encode = pd.get_dummies(X, columns  = ['model','transmission', 'fuelType',], drop_first = True)
```

```python
df.columns
```

```python
X_one_encode = X_one_encode.astype(int)
```

```python
X_one_encode
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.columns
```

```python
numeric_columns = ['year','mileage','tax','mpg', 'engineSize']
scaler = StandardScaler()
X_one_encode[numeric_columns] = scaler.fit_transform(X_one_encode[numeric_columns] )
```

```python
X_one_encode
```

```python
# Train-test split
from sklearn.model_selection import train_test_split

# Linear Regression model
from sklearn.linear_model import LinearRegression

# Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_one_encode, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
r2
```

```python
n = X_test.shape[0]

# Number of features
p = X_test.shape[1]

# Adjusted R²
adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
```

```python
adjusted_r2
```

```python
# Predictions
y_pred = model.predict(X_test)

# =========================
# Evaluation Metrics
# =========================

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Model Performance")
print("------------------")
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)


# =========================
# Sample Predictions
# =========================

print("\nSample Predictions")
print("-------------------")

for i in range(10):
    print(f"Actual: {y_test.iloc[i]} | Predicted: {y_pred[i]:.2f}")


# =========================
# Actual vs Predicted Plot
# =========================

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(y_test, y_pred, color='blue')

plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")

# Perfect prediction line
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    color='red'
)

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Store losses
loss_history = []

# Example: calculate loss for current prediction
y_pred = model.predict(X_train)

# Mean Squared Error Loss
loss = np.mean((y_train - y_pred) ** 2)

# Save loss
loss_history.append(loss)

print("Loss:", loss)
```

```python
plt.figure(figsize=(8,5))

plt.plot(loss_history, color='blue', marker='o')

plt.title("Loss Curve")
plt.xlabel("Iterations")
plt.ylabel("MSE Loss")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scores
scores = [r2, adjusted_r2]
labels = ['R²', 'Adjusted R²']

# Compact figure
plt.figure(figsize=(4,3))

bars = plt.bar(
    labels,
    scores,
    width=0.4,
    color=['royalblue', 'seagreen']
)

# Zoomed y-axis for compression
plt.ylim(min(scores) - 0.05, 1.0)

# Value labels
for bar in bars:
    yval = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval + 0.005,
        f'{yval:.3f}',
        ha='center',
        fontsize=9
    )

plt.title("R² Comparison", fontsize=11)

plt.tight_layout()
plt.show()
```
